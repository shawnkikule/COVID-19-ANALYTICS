```python
#Grouping data based on the WHO Region........
group_by=data.groupby('WHO Region')
x= group_by['Confirmed'].agg(np.mean)        #returns the average number of confirmed cases in each WHO region.....
q= group_by['Confirmed'].agg(np.sum)         #returns the total number of confirmed cases in each WHO Region......
y= group_by['Deaths'].agg(np.sum)            #returns the total number of deaths......
z= group_by['Active'].agg(np.sum)            #returns the total number of active cases........
p= group_by['Country'].agg(np.size)   #returns the total number of countries in each WHO Region.....
r =group_by['Recovered'].agg(np.sum)         #total number of recovered cases in each WHO Region......

print('')
print('The average number of cases confirmed in each WHO Region:')
print(x)
print('')
print('The total number of cases confirmed in each WHO Region:')
print(q)
print('')
print('The total number of deaths in each WHO Region:')
print(y)
print('')
print('The total number of active cases in each WHO Region:')
print(z)
print('')
print('The total number of countries in each WHO Region:')
print(p)
print('')
print('The total number of recovered cases in each WHO Region:')
print(r)
```

```python
#grouping by countries
grp1=data.groupby('Country')
x1= grp1['Confirmed'].agg(np.sum)
y1= grp1['Deaths'].agg(np.sum)
z1= grp1['Active'].agg(np.sum)
p1= grp1['Country'].agg(np.sum)
r1 =grp1['Recovered'].agg(np.sum)
print(x1)
print('')
print(y1)
print('')
print(p1)
print('')
print(r1)
```

```python
#Using a Heat Map to show the relationship between the columns and rows in the dataset.......
#These are a vizualization technique of representing data using colors.....
#Basically use color coding to replicate data representation.......
#use the seaborn import......

plt.figure(figsize=(10,10))
sbn.heatmap(data.corr(), annot=True, linewidths=0.05, fmt=' .2f', cmap='coolwarm')
plt.show()
```

```python
#scatter for confirmed and recovered
ty=data['Confirmed']
rt=data['Recovered']
plt.figure(figsize=(16,5))
plt.scatter(ty, rt, c='b')
plt.xlabel('Confirmed')
plt.ylabel('Recovered')
plt.show()
```

```python
#scatter for new cases and new recovered
q=data['New cases']
p=data['New recovered']
plt.figure(figsize=(16,5))
plt.scatter(q, p, c='r')
plt.xlabel('New cases')
plt.ylabel('New recovered')
plt.show()
```

```python
#Country/Region
plt.figure(figsize=(16,5))
plt.plot(p, 'ro', color='r')
plt.xticks(rotation=90)
plt.title('Countries/Regions and their respective Continents')
plt.xlabel('WHO Region')
plt.ylabel('No. of Countries/Regions')
plt.show()
```

```python
#Now that we've had a look at how the pandemic has affected the various countries in the Americas....

print('MOST AFFECTED COUNTRIES: -- UNITED STATES & BRAZIL')
print('Let us have a look at the recovery rates within this very region.')

import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,35)

x2 = data_america['Country']
y2 = data_america['Recovered']
      
plt.plot(y2,x2)
plt.title('Recovery rates from the COVID-19 pandemic among countries in the Americas region')
plt.ylabel('Countries in the Americas')
plt.xlabel("Cases of Recovery (per '1000 people)")
plt.grid(True)
plt.show()

```
